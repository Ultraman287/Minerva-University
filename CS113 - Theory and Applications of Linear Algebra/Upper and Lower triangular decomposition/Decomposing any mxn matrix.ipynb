```python
A = matrix([[1,-3,1],[2,-1,1],[6,-1,0]])
B = matrix(QQ, [[1,6,2],[2,12,5],[-1,-3,-1]])

def triangles(matrix):
    '''
    Creates an upper and lower triangle matrix from any given matrix.
    
    Inputs:
    matrix (matrix object in sage) : The original matrix which needs to be separated into upper and lower triangular forms
    
    Outputs:
    2 matrices: Desired Lower triangular matrix and Upper triangular matrix whose products are equal to the original matrix'''
   
    m = matrix.nrows()
    n = matrix.ncols()

    I = identity_matrix(QQ,m) 
    
    I_real = True
    
    #Here, nested for loops are implemented to go through the lower triangle within the matrix
    for a in range(len(matrix[0])-1):
        for b in range(a+1,matrix.nrows()): 
            if matrix[a,a] == 0: #Check to see whether a swap is needed
                for c in range(a+1,m): #Iterating to find the first non-zero row to swap with
                    if matrix[c,a] != 0:
                        matrix[c],matrix[a] = matrix[a],matrix[c]
                        I[c,a:],I[a,a:] = I[a,a:],I[c,a:]
                        I_real = False #Flag implemented to prevent premature updating during swap which results in the
                                        #wrong lower triangular matrix
                        break
            
            if I_real:
                I[b, a] = matrix[b, a]/matrix[a, a]
            I_real = True
            matrix[b] = matrix[b]-(matrix[a]/matrix[a, a])*matrix[b, a] 
            #Implementation of part 2 of the pseudocode where the specific relationship between the values is used to do row
            #operations that reduce the value to zero
            
            
            
            
    return I, matrix
            
        
```
