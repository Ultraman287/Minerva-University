````python
def fibonacci_iterative(n):
    """
    Uses iteration to output the nth number of the fibonacci sequence

    Parameters:
    n (int): Corresponds to the nth term of the fibonacci sequence

    Returns:
    int: Returns the value of the nth term

   """
    Current_number = 0
    Next_number = 1
    
    for i in range(n): #This iterator stores the current and next number which is calculated by adding on to previous sum
        Current_number, Next_number = Next_number, Current_number + Next_number
    
    return Current_number

def fibonacci_recursive(n):
    """
    Uses recursion to output the nth number of the fibonacci sequence

    Parameters:
    n (int): Corresponds to the nth term of the fibonacci sequence

    Returns:
    int: Returns the value of the nth term

   """
    
    if n == 0: #Two base cases defined
        return 0
    if n == 1:
        return 1
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2) #Using the recursive formulation of fibonacci

        
````

This sorting algorithm uses in-place sorting since it only needs a constant amount of space to store comparative variables and does the sorting on the original array itself rather than keeping a separate sorted list.

The process is similar to selection which tries to go through the entire array to find the minimum value. In this case, we try to find where in the array an item is supposed to go and directly swapping it into its intended index. This is done by comparing the item with the entire array and iterating a placeholder called 'intended position' for each value that's smaller than the item. This gives us the position we're supposed to switch to.

The following code block is another implementation of the above strategy where the item at the current index (i) is repeatedly sorted away to it's intended position until we end up with the right value at i. However, the while loop means that it is possible for an infinite loop which in this case happens in a list with duplicates because the index of the later duplicate will never be the same as the intended which is before it (due to the '<' condition).

The Algorithmic strategy this code uses is the concept of minimal swaps. Basically, for each item in the list, if you follow where the displaced number is supposed to be, you'll end up in a loop of numbers which end with the intended number at the index ending up back there. (Visualized below)

````python
def my_sort(array):
    """
    Sorts the given array in place
  
    Parameters:
    array (array): The list containing the numbers that need to be sorted
  
    Returns:
    array: list sorted in ascending order
  
    """
    
    # Iterating through the list
    for i in range(len(array)):
        
        # Placeholder value
        item = array[i]
        
        # Comparison done across the entire array to find intended position
        intended_position = 0
        for num in array:
            if num < item:
                intended_position += 1
                
        # Checking if the number is where it should be
        if i == intended_position:
            continue
        
        # Swapped into its intended position, number previously at intended made new item 
        array[intended_position], item = item, array[intended_position]
        
        # Getting the correct number to the position
        while i != intended_position:
            
            # Finding the intended position of the number currently at i
            intended_position = 0
            for num in array:
                if num < item:
                    intended_position += 1
            
            # Swapping that number out to its spot until the number (which is supposed to be at index i) is swapped in.
            array[intended_position], item = item, array[intended_position]
            
    return array
````

````python
def my_sort(old_array):
    """
    To solve the problem of duplicates, I decided to separate the duplicates initially and then later 
    insert them back into the array.
    I did it this way to preserve the minimum swap algorithmic strategy of the solution while not running into an infinite
    loop
    """
    
    duplicate_list = [] #Empty list for duplicates
    
    array = [] #Separate list without duplicates
    
    for i in range(len(old_array)): #Removing duplicates from the old array
        if old_array.count(old_array[i]) == 1:
            array.append(old_array[i])
        elif old_array.count(old_array[i]) > 1:
            duplicate_list.append(old_array[i])
            old_array[i] = float('inf')
    

    for i in range(len(array)):
        

        item = array[i]
        
        intended_position = 0
        for num in array:
            if num < item:
                intended_position += 1
                
        if i == intended_position:
            continue
        

        array[intended_position], item = item, array[intended_position]
        

        while i != intended_position:
            
            intended_position = 0
            for num in array:
                if num < item:
                    intended_position += 1
            
            array[intended_position], item = item, array[intended_position]
    
    for a in duplicate_list: #Adding the duplicates back into the list
        array.insert(array.index(a),a)
    
    return array
````

**(a)** Use the following pseudocode to implement `merge_3_sort()`. It is similar to merge sortonly that instead of splitting the initial array into halves, you will now split it into thirds, call the function recursively on each sublist, and then merge the triplets together. You might want to refer to this [beautiful resource](https://drive.google.com/file/d/1XH5bNiHhVchVWoCDLBtvjsCFkL9cU0do/view?pli=1) written by Prof. Drummond for details about the regular merge sort algorithm.

````python
def merge_3_sort(array, start, end):
    """
    Sorts array[start] to array[end] in place.
    E.g., to sort an array A, we will run 
    `merge_3_sort(A, 0, len(A)-1)`.


    Parameters
    ----------
    array : Python list or numpy array
    start : int
        index of array element to start sorting from
    end : int
        index of last array element to be sorted

    Returns
    -------
    array: a sorted Python list

    """
    
    if type(array) != list:
        print("Please put in an array")
        return
        
    
    if start  < end: #If statement that confirms whether the current list being sorted is not the base case (with a length of 1)
        q = start + (end-start)//3
        r = start + (2*(end-start))//3
        merge_3_sort(array,start,q)
        merge_3_sort(array,q+1,r) 
        merge_3_sort(array,r+1,end)
        merge_3(array,start,q,r,end) #calls upon the merge function defined above to fuse the divided (assumingly) sorted halves back together
        
    return array   
    
def merge_3(array, p, q, r, s):
    """
    Merges 3 sorted sublists 
    (array[p] to array[q], array[q+1] to array[r] and array[r+1] to array[s]) 
    in place.
    
    Parameters
    ----------
    array : Python list or numpy array
    p : int
        index of first element of first sublist
    q : int
        index of last element of first sublist
    r : int
        index of last element of second sublist
    s : int
        index of last element of third sublist

    """
    Left = array[p:q+1] #Creating the 3 divisions
    Middle = array[q+1:r+1] 
    Right = array[r+1:s+1] 
    
    Left.append(float('inf')) #This adds the sentinels that act as a placeholder for the end of one list 
    Middle.append(float('inf'))
    Right.append(float('inf'))
    
    i = j = k = 0 #Initializing the iterators i and j
   
    for current_index in range(p,s+1): #Iterating through the subarray within A that will be reordered
        if Left[i] <= Middle[j] and Left[i] <= Right[k] : 
            array[current_index] = Left[i]
            i += 1 
        elif Middle[j] <= Right[k] and Middle[j] <= Left[i] : 
            array[current_index] = Middle[j]
            j += 1
        else:
            array[current_index] = Right[k]
            k += 1
  
    return array
````

### Analytical solution ###

From the working above, we can see the relative orders of growth for merge_3_sort(BigO(log3(n) * n)) and selection sort (BigO(n^2)). This tells us that no matter what, after a certain input size, selection sort will always take more time than merge_3_sort since it has a higher order of growth.
However, in real world term, this point of inflection is seen with large array sizes due to the greater amount of operations and space that merge_3_sort would need for even the smaller inputs. Thus, if we can find the point of inflection graphically, we can find out exactly where it would be less expensive to implement selection sort instead.

### Graphical solution ###

To find that inflection point, the method I used was to use the worst case scenario arrays for selection sort of increasing sizes (reversed lists) and finding the amount of time it takes to sort the array for both selection and merge_3_sort. From there, I was able to find a general pattern for growth at the lower levels for merge_3_sort (by repetitively making graphs over the input size) and use that along with the selection sort times to find the point where it would be less costly to use selection.

````python
#Analysis of time it takes to run merge sort over different input sizes

import timeit
import matplotlib.pyplot as plt

def selection_sort(A):
    """
    Sorts the given array in place by finding the minimum value in the rest of the list at each iteration and swapping it
    into its position.
  
    Parameters:
    array (array): The list containing the numbers that need to be sorted
  
    Returns:
    array: list sorted in ascending order
  
    """
    
    
    n = len(A)

    for i in range(0, n): #iterating through the values up to but not including the final item in the list
        minidx = i
        for j in range(i+1, n): #finding the smallest number in the list after the sorted part
            if A[j] < A[minidx]:
                minidx = j
        A[i], A[minidx] = A[minidx], A[i]
    
   
    
    return A

def selection_sort_time(A):
    """
    Gives the time required to run the function
  
    """
    Start = timeit.default_timer()
    
    n = len(A)

    for i in range(0, n): #iterating through the values up to but not including the final item in the list
        minidx = i
        for j in range(i+1, n): #finding the smallest number in the list after the sorted part
            if A[j] < A[minidx]:
                minidx = j
        A[i], A[minidx] = A[minidx], A[i]
    
    Stop = timeit.default_timer()
    
    return Stop - Start

for a in range(3):    #creating the three lists between selection and merge
    for a in range(10):

        testing = []

        testing_b = []

        for a in range(1,75):
            testing.append(list(range(a,0,-1)))
            testing_b.append(list(range(a,0,-1)))



        testing_times = []
        selection_testing_times = []


        for a in testing:
            Start = timeit.default_timer()
            merge_3_sort(a, 0, len(a)-1)
            Stop = timeit.default_timer()
            testing_times.append(Stop-Start)

        for a in testing_b:
            selection_testing_times.append(selection_sort_time(a))


        x_values = [i for i in range(1,75)]

        plt.plot(x_values, testing_times, alpha = 0.4) #many different for merge sort to get an approximate average
    plt.plot(x_values, selection_testing_times)
    plt.xlabel('Size of reversed array')
    plt.ylabel('Time taken')
    plt.axvline(x=65)
    plt.show()
````

Looking at the three graphs above us, we can clearly see the general evolution of the times of the two algorithms with the curved one below in all three cases showing selection sort. In all 3 graphs, the inflection point occurs around 65 which tells us that in any case where the array's size is less than 65, we should use selection sort instead of merge sort. However, in the implementation, it will take some time to add in the selection sort part so we can be safer with less than 60.

````python
def merge_3_sort_threshold(array, start, end):
    """
    Sorts the given subarray with a mix of merge and selection sort where selection sort is used when the length of the array
    is less than 60
  
    Parameters:
    array (array): The list containing the numbers that need to be sorted
  
    Returns:
    array: list sorted in ascending order
  
    """
    
    if type(array) != list:
        print("Please input a list")
        return
        
    if end - start < 60: #Condition for applying selection sort
        to_sort = array[start:end+1]
        selection_sort(to_sort) #selection sort done on a dummy list
        
        mini_iterator = 0 #to iterate through the same points on the main array and the sorted dummy array
        for i in range(start,end+1):
            array[i] = to_sort[mini_iterator] #transferring the sorted subarray back in
            mini_iterator += 1
        
        return array
    
    if start  < end: 
        q = start + (end-start)//3
        r = start + (2*(end-start))//3
        merge_3_sort_threshold(array,start,q)
        merge_3_sort_threshold(array,q+1,r) 
        merge_3_sort_threshold(array,r+1,end)
        merge_3(array,start,q,r,end) 
        
    return array  
    
def merge_3(array, p, q, r, s):
    Left = array[p:q+1] #Creating the 3 divisions
    Middle = array[q+1:r+1] 
    Right = array[r+1:s+1] 
    
    Left.append(float('inf')) #This adds the sentinels that act as a placeholder for the end of one list 
    Middle.append(float('inf'))
    Right.append(float('inf'))
    
    i = j = k = 0 #Initializing the iterators i and j
   
    for current_index in range(p,s+1): #Iterating through the subarray within A that will be reordered
        if Left[i] <= Middle[j] and Left[i] <= Right[k] : 
            array[current_index] = Left[i]
            i += 1 
        elif Middle[j] <= Right[k] and Middle[j] <= Left[i] : 
            array[current_index] = Middle[j]
            j += 1
        else:
            array[current_index] = Right[k]
            k += 1
  
    return array
````
